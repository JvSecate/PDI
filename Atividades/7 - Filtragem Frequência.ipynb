##1.     Calcule e visualize o espectro de uma imagem 512x512 pixels:

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

####a)   crie e visualize uma imagem simples – quadrado branco sobre fundo preto;

```python
img = np.zeros((512, 512), dtype=np.uint8)
img[200:300, 200:300] = 255
```

####b)   calcular e visualizar seu espectro de Fourier (amplitudes);

```python
f_transform = np.fft.fft2(img)
f_shifted = np.fft.fftshift(f_transform)
magnitude_spectrum = np.abs(f_shifted)
phase_spectrum = np.angle(f_shifted)
```

####c)    calcular e visualizar seu espectro de Fourier (fases);

```python
center = (img.shape[1] // 2, img.shape[0] // 2)
angle = 40
scale = 1.0
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
```

####d)   obter e visualizar seu espectro de Fourier centralizado;

```python
img_rotated = cv2.warpAffine(img, rotation_matrix, (img.shape[1],
img.shape[0]))
f_transform_rotated = np.fft.fft2(img_rotated)
f_shifted_rotated = np.fft.fftshift(f_transform_rotated)
magnitude_spectrum_rotated = np.abs(f_shifted_rotated)
```

####e)   Aplique uma rotação de 40º no quadrado e repita os passo b-d;

```python
phase_spectrum_rotated = np.angle(f_shifted_rotated)
```

####f)   Aplique uma translação nos eixos x e y no quadrado e repita os passo b-d;

```python
M_translation = np.float32([[1, 0, 50], [0, 1, 50]])
img_translated = cv2.warpAffine(img, M_translation, (img.shape[1],
img.shape[0]))
```

####g)   Aplique um zoom na imagem e repita os passo b-d;

```python
f_transform_translated = np.fft.fft2(img_translated)
f_shifted_translated = np.fft.fftshift(f_transform_translated)
magnitude_spectrum_translated = np.abs(f_shifted_translated)
```

```python
phase_spectrum_translated = np.angle(f_shifted_translated)
img_zoomed = cv2.resize(img, None, fx=1.5, fy=1.5,
interpolation=cv2.INTER_LINEAR)
f_transform_zoomed = np.fft.fft2(img_zoomed)
f_shifted_zoomed = np.fft.fftshift(f_transform_zoomed)
magnitude_spectrum_zoomed = np.abs(f_shifted_zoomed)
phase_spectrum_zoomed = np.angle(f_shifted_zoomed)
fig, ax = plt.subplots(nrows=1, ncols=6, figsize=(16, 8))

ax[0].imshow(img, cmap='gray')
ax[0].set_title("Imagem original")
ax[0].axis('off')
ax[1].imshow(np.log(1 + magnitude_spectrum), "gray")
ax[1].set_title("Amplitude")
ax[1].axis('off')
ax[2].imshow(phase_spectrum, "gray")
ax[2].set_title("Fase")
ax[2].axis('off')
ax[3].imshow(np.log(1 + magnitude_spectrum), "gray")
ax[3].set_title("Espectro centralizado")
ax[3].axis('off')
ax[4].imshow(np.log(1 + magnitude_spectrum), "gray")
ax[4].set_title("Espectro amplitude")
ax[4].axis('off')
ax[5].imshow(np.log(1 + magnitude_spectrum_translated), "gray")
ax[5].set_title("Espectro rotacionado")
ax[5].axis('off')

fig, ax = plt.subplots(nrows=1, ncols=5, figsize=(16, 8))

ax[0].imshow(phase_spectrum_rotated, "gray")
ax[0].set_title("Fase rotacionada")
ax[0].axis('off')
ax[1].imshow(np.log(1 + magnitude_spectrum_translated), "gray")
ax[1].set_title("Espectro transladada")
ax[1].axis('off')
ax[2].imshow(phase_spectrum_translated, "gray")
ax[2].set_title("Fase transladada")
ax[2].axis('off')
ax[3].imshow(np.log(1 + magnitude_spectrum_zoomed), "gray")
ax[3].set_title("Espectro zoom")
ax[3].axis('off')
ax[4].imshow(phase_spectrum_zoomed, "gray")
ax[4].set_title("Fase zoom")
ax[4].axis('off')
```

```python
def compute_spectrum(img):
 f_transform = np.fft.fft2(img)
 f_shifted = np.fft.fftshift(f_transform)
 magnitude_spectrum = np.abs(f_shifted)
 return np.log(1 + magnitude_spectrum)
```

####b)    a imagem do spectro de fourier

```python
def ideal_lowpass_filter(shape, cutoff):
 rows, cols = shape
 center_row, center_col = rows // 2, cols // 2
 y, x = np.fft.fftfreq(rows), np.fft.fftfreq(cols)
 X, Y = np.meshgrid(x, y)
 D = np.sqrt(X**2 + Y**2)
 return np.where(D <= cutoff, 1, 0)
```

####c)    a imagem de cada filtro;

```python
def butterworth_lowpass_filter(shape, cutoff, order=2):
 rows, cols = shape
 center_row, center_col = rows // 2, cols // 2
 y, x = np.fft.fftfreq(rows), np.fft.fftfreq(cols)
 X, Y = np.meshgrid(x, y)
 D = np.sqrt(X**2 + Y**2)
 return 1 / (1 + (D / cutoff)**(2 * order))
```

####d)     a imagem resultante após aplicação de cada filtro.

##5.     Efetue o mesmo que se pede no item 4, mas use o filtro passa-alta em vez do filtro passa-baixa.

```python
def show_images(images, titles, rows=1, cols=3):
 fig, axes = plt.subplots(rows, cols, figsize=(15, 12))
 for i, ax in enumerate(axes.flat):
 ax.imshow(images[i], cmap='gray')
 ax.set_title(titles[i])
 ax.axis('off')
 plt.tight_layout()
 plt.show()
def ideal_bandpass_filter(shape, low_cutoff, high_cutoff):
 rows, cols = shape
 y, x = np.fft.fftfreq(rows), np.fft.fftfreq(cols)
 X, Y = np.meshgrid(x, y)
 D = np.sqrt(X**2 + Y**2)
 return np.where((D >= low_cutoff) & (D <= high_cutoff), 1, 0)
def butterworth_bandpass_filter(shape, low_cutoff, high_cutoff,
order=2):
 rows, cols = shape
 y, x = np.fft.fftfreq(rows), np.fft.fftfreq(cols)
 X, Y = np.meshgrid(x, y)
 D = np.sqrt(X**2 + Y**2)
 return 1 / (1 + ((D * (D < high_cutoff) * (D > low_cutoff)) ** (-2
* order)))
def gaussian_bandpass_filter(shape, low_cutoff, high_cutoff):
 rows, cols = shape
 y, x = np.fft.fftfreq(rows), np.fft.fftfreq(cols)
 X, Y = np.meshgrid(x, y)
 D = np.sqrt(X**2 + Y**2)
 return np.exp(-((D - (low_cutoff + high_cutoff) / 2) ** 2) / (2 *((high_cutoff - low_cutoff) ** 2)))
```
